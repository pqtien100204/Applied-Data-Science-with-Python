```python
import pandas as pd
import numpy as np
from scipy import stats
import datetime
```

Hypothesis Testing is a type of statistical analysis in which you put your assumptions about a population parameter to the test. It is used to estimate the relationship between 2 statistical variables.

Statistical Hypothesis testing: The null hypothesis of a test always predicts no effect or no relationship between variables, while the alternative hypothesis states your research prediction of an effect or relationship.

```python
csv = pd.read_csv("grades.csv")
df = pd.DataFrame(csv)
df.head()
```

```python
# Data exploration to understand this dataset more
shpe = df.shape
print(f"There are {shpe[0]} rows, {shpe[1]} columns")
print(df.dtypes)
```

```python
# Task 1: Data Segmentation / Categorization => student finished
# assignment before December 2015 => early finisher
# after that => late finisher
pd.to_datetime(df["assignment1_submission"])
earler = df[df["assignment1_submission"] < "2016"]
# lateer = df[df["assignment1_submission"] > "2016"]
```

```python
earler.head()
```

```python
lateer = df[~df.index.isin(earler.index)]
lateer.head()
```

```python
# use the .apply() function to add the row of early and late finishers
def submission(x):
    if x["assignment1_submission"] < "2016":
        return "Early"
    else:
        return "Late"
    
df["Submission"]=df.apply(submission, axis="columns")
df.head()
```

```python
print(earler["assignment1_grade"].mean())
print(lateer["assignment1_grade"].mean())
```

```python
from scipy.stats import ttest_ind
ttest_ind(earler["assignment1_grade"],lateer["assignment1_grade"])
```

```python
# Let's set the significance level to be 0.05 or 5% 
# we can see that the probability level is 0.18
# pvalue > pre-defined threshold => cannot reject the null hyp (statistically insignificant)
# => the nul hyp is that the 2 populations are the same 
# But we still don't have enough evidence for the similarity between 2 datasetsb 
```

```python
# Create a dataset with 10000 datapoints
df1 = pd.DataFrame([np.random.random(100) for x in range(100)])
df1.head()
```

```python
df2 = pd.DataFrame([np.random.random(100) for x in range(100)])
df2.head()
```

```python
ttest_ind(df1,df2)
```

The p value reports the probability we can get the observed data under the null hypothesis, so with a larger p value (by larger we mean it is closer to 1, instead of saying that it is more extreme), we have less evidence to reject the null hypothesis.

With a larger alpha, we reject the null hypothesis less carefully.

Significance means the evidence we observed from the data against the null hypothesis, and alpha level is a measure of our tolerance of the significance.
